```python
from sympy import *
```

```python
A, B, C, D, E, F = symbols("A, B, C, D, E, F", complex=True)
z = symbols("z", complex=True, nonzero=True)
w = symbols("w", real=True, positive=True, nonzero=True)
zbar = symbols("\\bar{z}", complex=True, nonzero=True)
wbar = symbols("\\bar{w}", real=True, positive=True, nonzero=True)
x = symbols("x", real=True)
k, beta = symbols("k, \\beta", real=True, positive=True)
a = symbols("a", real=True, positive=True, nonzero=True)
# z = exp(I*k*a)
# w = exp(beta*a)

psi1 = (z + B/z).subs(a, x)
psi2 = (C*w + D/w).subs(a, x)
psi3 = (E*z).subs(a, x)
```

```python
eq1 = zbar + B*z - C*wbar - D*w
eq2 = I*k*zbar - I*k*B*z - beta*C*wbar + beta*D*w
eq3 = C*w + D*wbar - E*z
eq4 = beta*C*w - beta*D*wbar - I*k*E*z
```

```python
soln = solve([eq1, eq2, eq3, eq4], [B, C, D, E], dict=True)[0]
soln
```
