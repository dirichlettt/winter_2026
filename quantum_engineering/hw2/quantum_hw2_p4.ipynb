```python
from sympy import *
```

```python
h, a, C, m, w = symbols(r"h, a, C_0, m, \omega", real=True, positive=True)
x = symbols("x", real=True)
```

```python
psi = C*exp(-a*x**2)
```

```python
psi*diff(psi, x, 2)
```

```python
2*-h**2*integrate(4*a**2*x**2*exp(-2*a*x**2), (x, 0, oo))
```

```python
2*-h**2*integrate(-2*a*exp(-2*a*x**2), (x, 0, oo))
```

```python
integrate(psi*diff(psi, x, 1), (x, 0, oo)) + integrate(psi*diff(psi, x, 1), (x, oo, 0))
```

```python
integrate(x*psi**2, (x, 0, oo)) + integrate(x*psi**2, (x, -oo, 0))
```

```python
expected_x_sq = integrate(x**2*psi**2, (x, 0, oo)) + integrate(x**2*psi**2, (x, -oo, 0))
expected_x_sq
```

```python
expected_p_sq = -h**2*integrate(psi*diff(psi, x, 2), (x, 0, oo))
2*expected_p_sq
```

```python
sqrt(expected_x_sq)
```

```python
sqrt(expected_p_sq)
```

```python
heisenberg = sqrt(expected_x_sq) * sqrt(expected_p_sq)
heisenberg
```

```python
heisenberg.subs(C, Pow(m*w/(pi*h), Rational(1,4))).subs(a, m*w/(2*h))
```
